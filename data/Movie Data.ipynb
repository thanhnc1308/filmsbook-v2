```python
# get all genres

import requests
import json

api_key = 'eb3cab35c6e825e5209fbf847b4d2e40'
base_url = 'https://api.themoviedb.org/3/genre/movie/list'
url = base_url + '?api_key=' + api_key + '&language=en-US'

r = requests.get(url = url)
genres = r.json()['genres']
for genre in genres:
    obj = {
        'moviedb_id': genre['id'],
        'name': genre['name']
    }
    requests.post('http://localhost/filmsbook-v2/genres/store', data=obj)
```
